```python
# A function to find the string entries in the column 'YearsCode' and 'YearsCodePro'
def FindStringEntry(df, column):
    strings=set()
    for value in df[column]:
        try:
            float(value)
        except:
            strings.add(value)
    return strings

YearsCode_Strings= FindStringEntry(df, 'YearsCode')
YearsCodePro_String= FindStringEntry(df, 'YearsCodePro')

YearsCode_Strings, YearsCodePro_String
```

```python
def FindNotInt(df, column):
    count=0
    for value in df[column]:
        try:
            float(value)
        except ValueError:
            count+=1
             
    return count

YearsCodeNonInt=FindNotInt(df, 'YearsCode')
YearsCodeProNonInt=FindNotInt(df, 'YearsCodePro')

YearsCodeNonInt, YearsCodeProNonInt
```

<span style="color:red; font-family:monospace;">Cleaning Columns 'YearsCode' and 'YearsCodePro'</span>

```python
import numpy as np
Remove = {'Less than 1 year', 'More than 50 years'}

df['YearsCode'] = df['YearsCode'].apply(lambda x: np.nan if x in Remove else x)
df['YearsCodePro'] = df['YearsCodePro'].apply(lambda x: np.nan if x in Remove else x)
```

```python
import numpy as np

def cleanAge(value):
    if value == 'Prefer not to say': 
        return np.nan  
    elif 'Under 18 years old' in value: 
        return '0-17' 
    elif '65 years or older' in value: 
        return '65-80'  
    else:
        return value.split(' years old')[0]  

df['Age'] = df['Age'].apply(cleanAge)
df['Age'].head(10)
```
